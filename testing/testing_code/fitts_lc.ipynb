```python

time = np.zeros(len(data))
rate = np.zeros((len(data)))#, 4))
error = np.zeros((len(data)))#, 4))
totcounts =  np.zeros(len(data))

for i in range(len(data)):
    time[i] = data[i][0]
    rate[i] = data[i][1]
    error[i] = data[i][2]
    totcounts[i] = data[i][3]
```
